```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = numpy.arange(40,50)
b = numpy.arange(50,60)
plt.title("Line Chart", color = "b")
plt.xlabel("x-Axis", color = "b")
plt.ylabel("y-Axis", color = "b")
plt.plot(a, b)
```

```python
x = [1,2,3,4,5,6,7] #days of the week
y = [160,150,140,145,175,165,180] #Sales
plt.title("Sales per Week", color = "b")
plt.xlabel("Days of week", color = "b")
plt.ylabel("Sales", color = "b")
plt.plot(x,y)
```

```python
days = [1,2,3,4,5,6,7]
sales_1 = [160, 150, 140, 145, 175, 165, 180]
sales_2 = [70, 90, 160, 150, 140, 145, 175]
plt.plot(days,sales_1)
plt.plot(days,sales_2)
plt.title("Chart of Sales per Week")
plt.xlabel("Days of week", color = "b")
plt.ylabel("Sales", color = "b")
```
